# RNN Optimizer Comparison
Training RNN on temperature dataset with different optimizers

```python
from examples.train_sydney_dataset import train_model
```

```python
optimizers = ["sgd", "momentum", "rmsprop", "adam", "adagrad"]

all_losses = {}

for opt in optimizers:
    print(f"\nTraining with {opt.upper()}")
    losses = train_model(
        optimizer_name=opt,
        epochs=50,
        batch_size=32,
        hidden_size=64,
        seq_len=30,
        verbose=True
    )
    all_losses[opt] = losses
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for opt, losses in all_losses.items():
    plt.plot(losses, label=opt)

plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Optimizer Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
